```python
""" !sudo conda install -c conda-forge folium """
import folium
import requests
import pandas as pd
import numpy as np


arrest_table = pd.read_csv("BPD_Arrests.csv", parse_dates=[['arrestDate', 'arrestTime']])

arrest_df = arrest_table[pd.notnull(arrest_table["Location 1"])]
arrest_df["lat"], arrest_df["long"] = arrest_table["Location 1"].str.split(",").str
arrest_df["lat"] = arrest_df["lat"].str.replace("(", "").astype(float)
arrest_df["long"] = arrest_df["long"].str.replace(")", "").astype(float)
#arrest_df.head()
```

```python
grouper = pd.Grouper(freq='W', base=0, key='arrestDate_arrestTime')
#arrest_dt_lists = list(gp[["lat","long"]].values.tolist() for idx, gp in arrest_df.groupby(grouper))
arrest_dt_lists = list(gp[["lat","long"]].values.tolist() for idx, gp in arrest_df.groupby(arrest_df["arrestDate_arrestTime"].dt.dayofweek))
```

```python
from folium import plugins
from folium.plugins import HeatMap

map = folium.Map(location=[39.29, -76.61], zoom_start=11, zoom_control=False, scrollWheelZoom=False,dragging=False)

hm = plugins.HeatMapWithTime(arrest_dt_lists, auto_play=True,max_opacity=0.8, radius = 3, gradient={.4: 'blue', .65: 'lime', 1: 'red'})
hm.add_to(map)
map
```

```python
# This map plots a heatmap of arrests on each day of the week starting on Monday.  The map displays that the number of arrests increases through the middle of the week to its peak, then slowly decreases throughout the week.  There is then a sharp drop on Sunday with a relatively small number of arrests.
```
